<h2>Literature Review</h2>
<div class='container' style='font-size:20px'>
    <p>In the course of our project, we reviewed several articles related to BHT and BHP predictions. These articles provided valuable insights and methodologies that guided our approach to predicting BHT and BHP.
    </p>
    <p>
        The first article, ”A New Model for Predicting Bottomhole Temperatures During Drilling Operations” by Al-Khdheeawi et al. (2020), presents a new model for predicting BHT during drilling operations. The model is based on real-time data and considers various factors such as drilling fluid properties, circulation rate, and depth. The authors tested their model on several wells and found it to be more accurate than existing models [1].
    </p>
    <p>
        The second article, ”Spatial Prediction for Bottom Hole Temperature and Geothermal Gradient in Colombia” by Matiz-Le´on, provides a comprehensive study on the spatial prediction of BHT and geothermal gradient in Colombia. The study used deterministic methods of minimum curvature and IDW, and probabilistic methods using Kriging and Sequential Gaussian Simulation (SGS). The study found that the relative variance contains fluctuations when the spatial prediction is performed by Kriging and SGS [2].
    </p>
    <p>
        The third article, ”Effect of temperature variation on the accurate prediction of bottom-hole pressure in well drilling” by Ebrahim Hajidavalloo, Alireza Daneh-Dezfuli, and Ali Falavand Jozaei, focuses on the impact of various parameters on drilling fluid, which is a crucial part of a successful and safe drilling operation. The study uses the Bingham-plastic model to simulate drilling operations and develop temperature and pressure distribution throughout the wellbore. The research explores the effects of changes in drilling fluid flow rate, drilling fluid specific heat, and inlet mud temperature for an oil-based mud. It also considers the effects of frictional pressure loss, pressure loss in the drill bit, and drill string rotation as internal heat sources in the wellbore. The results indicate that thermal interaction is necessary for the accurate prediction of the bottom-hole pressure. Variations in different parameters alter the temperature distribution in the wellbore, which in turn changes the bottom-hole pressure. The study found that considering the variation of rheological properties, bottom-hole pressure prediction may change as much as 685 psi (about 2.30%), which is significant. The model’s predictions were compared with field data for validation [3].
    </p>
    <p>
        The fourth article, titled ”Bottom Hole Pressure Estimation using Hybridization Neural Networks and Grey Wolves Optimization,” Amar, Zeraibi, and Redouane address the significance of bottom hole pressure (BHP) in production and reservoir studies. BHP plays a crucial role in well efficiency, completion design, and overall well strategies. However, traditional methods for determining BHP, such as pressure gauges, present limitations in terms of cost and handling of noisy data. To overcome these limitations, the authors propose a novel approach for estimating BHP in vertical wells with multi-phase flow. They employ an Artificial Neural Network (ANN) based on back-propagation training (BP-ANN) with a hidden layer consisting of 12 neurons, determined through trial and error. Furthermore, they incorporate Grey Wolves optimization to fine-tune the weights and thresholds of the neural network. To evaluate the accuracy of their model, the researchers collected 100 field data points from Algerian fields, covering a wide range of variables. They compared the performance of their proposed model with Genetic Algorithm (GA) and Particle Swarm Optimization (PSO), two commonly used optimization algorithms inspired by natural phenomena. The results demonstrated the superior accuracy of the proposed model when compared to other studies, as indicated in the provided table. This research contributes to the advancement of BHP estimation by offering a more accurate and cost-effective approach, surpassing the limitations of traditional methods and enhancing the understanding and optimization of well operations in various field conditions [4].
    </p>
    <p>
        The fifth article, Tariq, Mahmoud, and Abdulraheem propose a computational method that outperforms conventional Artificial Neural Network (ANN) models for estimating bottom-hole pressure. They achieve an impressive accuracy, with an absolute percentage error of less than 2.1 percent using their model.To develop their model, the authors utilize real-time surface production data. They employ the Particle Swarm Optimization (PSO) algorithm to optimize the weights and biases of the trained neural network. The training dataset consists of 206 data points, obtained from reliable sources such as Govier and Fogarasi (1975) and Asheim (1986). These data points were collected by lowering a pressure gauge into the well, and they include relevant parameters such as depth, flow rates, tubing ID, API gravity, well-head pressure, bottom hole temperature (BHT), and bottom hole pressure (BHP). By employing their hybrid neural network approach and leveraging the PSO algorithm, the authors achieve significant improvements in bottom-hole pressure estimation compared to traditional ANN models. This research contributes to the advancement of accurate real-time predictions in the field of oil and gas production [5].
    </p>

<div style="text-align:center"><header><h1>3.3. Creating New Columns</h1></header></div>

<div>
<p>This section involves splitting the column named <strong>DATEPRD</strong>, which contains datetime values, into complete date and year-only columns. This will allow for a more convenient plotting in subsequent sections.</p>
<p>In addition, the dataset will be grouped by <strong>Years</strong> and the <strong>NPD_WELL_BORE_NAME</strong> features to compare the observations for each well with respect to the corresponding years.</p>
<p>Finally, cumulative production values for each phases will be calculated, and named as folows:</p>

<ul>
<li><i>BORE_OIL_CUM</i>, represents cumulative oil production</li>
<li><i>BORE_GAS_CUM</i>, represents cumulative gas productions</li>
<li><i>BORE_WAT_CUM</i>, represents cumulative water production</li>
</ul>
</div>
</body>
</html>

```python
years = volve_production_producing["DATEPRD"].apply(lambda x: x.strftime("%Y"))

#Implementing the new columns,
volve_production_producing.insert(0, 'Years', years)

#Groupping the data with respect to the years,
annual_volve_production = volve_production_producing.groupby(by=['Years', 'NPD_WELL_BORE_NAME'], as_index=True)[['BORE_OIL_VOL', 'BORE_GAS_VOL', 'BORE_WAT_VOL']].sum()
```

```python
#Creating cumulative production columns for each phases,
volve_production_producing.loc[:, "BORE_CUM_OIL"] = volve_production_producing.loc[:, 'BORE_OIL_VOL'].cumsum() 
volve_production_producing.loc[:, "BORE_CUM_GAS"] = volve_production_producing.loc[:, 'BORE_GAS_VOL'].cumsum()
volve_production_producing.loc[:, "BORE_CUM_WAT"] = volve_production_producing.loc[:, 'BORE_WAT_VOL'].cumsum()

volve_production_producing.loc[:, "TOT_CUM_PRO"] = volve_production_producing.loc[:, "BORE_CUM_OIL"] + volve_production_producing.loc[:, "BORE_CUM_GAS"] + volve_production_producing.loc[:, "BORE_CUM_WAT"]

#Results,
volve_production_producing.head(10)
```

<p>In this particular section, we will showcase the production quantities of each phase as well as the total production. This will allow for a visually striking depiction of the distribution of each fluid's production share.</p>
<p>Furthermore, we will create separate plots for the production quantities of each fluid, highlighting their respective peak amounts to illustrate the decline curves.
```python
# Calculating the total Oil, Gas, and Water produced
total_oil = volve_production_producing.BORE_OIL_VOL.sum()
total_gas = volve_production_producing.BORE_GAS_VOL.sum()
total_water = volve_production_producing.BORE_WAT_VOL.sum()

three_phase_production = total_oil + total_gas + total_water
```

```python
print(f"The total oil produced is {total_oil} cubic meters")
print(f"The total gas produced is {total_gas} cubic meters")
print(f"The total water produced is {total_water} cubic meters")
print('---------------------------------------------------------')
print(f"The total amount produced is {round(three_phase_production)} cubic meters, {round(three_phase_production * 6.28981077)} barrels.")
```

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
```

```python
#We are going to create a pie chart that contains 3 phase production with respect to their production amounts. To do so,
phases = ['Oil', 'Gas', 'Water']
phases_amounts = [round(total_oil), round(total_gas), round(total_water)]

fig = px.pie(values=phases_amounts, names=phases, labels=phases, color=phases,
             color_discrete_map={'Oil':'green', 'Gas':'red', 'Water':'blue'},
             title="<b>Fluid Production</b>")

fig.update_traces(hovertemplate="<br>".join(["Fluid: %{label}", "Production: %{value} standard meter cube"]),
                 pull=[0.3, 0.4, 0.3],
                 textinfo='percent')

fig.update_layout(font_size=16)
fig.show()
```

```python
annual_oil_total = annual_volve_production.groupby(by=["Years"])[["BORE_OIL_VOL"]].sum()


#Annual Oil Production Plot, with curve fitting
layout = go.Layout(title='<b>Annual Oil Production</b>',
                   xaxis_title='Year',
                   yaxis_title='Oil Production (meters cube)',
                   height=500,
                   width=900,
                   font_size=12)


figure = go.Figure(layout=layout)

figure.add_traces(data=go.Bar(x=annual_volve_production.index.get_level_values('Years') , y=annual_volve_production["BORE_OIL_VOL"],
                           marker=dict(color='limegreen', line=dict(color='white', width=1.2)),
                           opacity=.8,  
                           hovertemplate='<br>'.join(['Year = %{x}', 'Oil = %{y}']),
                           name='Production'))

figure.add_traces(data=go.Scatter(x=annual_oil_total.index, y=annual_oil_total["BORE_OIL_VOL"],
                               line=dict(dash='dash', color='red', width=3),
                               marker_size=10,
                               hovertemplate='<br>'.join(['Year = %{x}', 'Oil = %{y}']),
                               name='Peaks'))

figure.show()
```

```python
#To be able to get proper plot, we have change and sort index only by Years.
annual_gas_total = annual_volve_production.groupby(by=["Years"])[["BORE_GAS_VOL"]].sum()

#Annual Gas Production Plot, with curve fitting
layout = go.Layout(title='<b>Annual Gas Production</b>',
                   xaxis_title='Year',
                   yaxis_title='Gas Production (Cubic Meters)',
                   height=500,
                   width=900,
                   font_size=12)


figure = go.Figure(layout=layout)

figure.add_traces(data=go.Bar(x=annual_volve_production.index.get_level_values('Years') , y=annual_volve_production["BORE_GAS_VOL"],
                           marker=dict(color='indianred', line=dict(color='white', width=1.2)),
                           opacity=.8,   
                           hovertemplate='<br>'.join(['Year = %{x}', 'Gas = %{y}']),
                           name='Production'))

figure.add_traces(data=go.Scatter(x=annual_gas_total.index, y=annual_gas_total["BORE_GAS_VOL"],
                               line=dict(dash='dash', color='deepskyblue', width=3),
                               marker_size=10,
                               hovertemplate='<br>'.join(['Year = %{x}', 'Gas = %{y}']),
                               name='Peaks'))

figure.show()
```

<div style="text-align:center"><header><h1>5. Data Imputation</h1></header></div>

<div class='container' style='mt-2'>
<p>The impementation of a successful data imputation plays a tremendous affect in the data science world, as the wrong imputations may result huge deviations from the real data. As a results of that, the complete analysis, and the corresponding model will be useless.</p>
<p>To avoid from that occasion, we will be conducting some test to identify the types of missinges of each feature which contains <code>NaN</code> values. Our methodology will based on the followings:</p>
    
<ul>
<li>
<code>msno</code> matrix plot; which demonstrates the missing values within the each feature. By looking at the <code>msno</code> matrix plot, one may identify a possible correlation between missingess of features as the dependent feature's observations will be lost at the same time. 
</li>
<li>
<strong>Missing Completely Random Test;</strong> involves <code>Independent T-Test</code>.
</li>
<li>
<strong>Missing at Not Random Test;</strong> involves a simple <code>correlation test</code>.
</li>
<li>
<strong>Missing at Random;</strong> involves <code>KNNImputer test</code>.
</li>
</ul>
</div>

<hr>
<br>
<div class='card-mt-4'>
    <h3>Independent T-Test</h3>
    <p>The independent t-test compares the means of the two groups and calculates a t-statistic, which measures the difference between the means relative to the variability within the groups. The t-statistic is then compared to a critical value from the t-distribution to determine the statistical significance.</p>
    <p>The null hypothesis of the independent t-test assumes that there is no significant difference between the means of the two groups, while the alternative hypothesis suggests that there is a significant difference.</p>
    <p>In our case, the null hypothesis will be features are independent to each other in a means of missingness type.</p>
    <p>The t-statistic is calculated as the difference between the sample means divided by the standard error of the difference. A smaller p-value associated with the t-statistic indicates stronger evidence against the null hypothesis, suggesting a significant difference between the groups.</p>
</div>
<br>
<div class='card-mt-4'>
    <h3>Correlation Test</h3>
    <p>It simply looks to the possible direct correlations within the missing values, and gives features' correlation coefficients between -1 and 1. 1 indicates strong correlation. That is, the resulted features can be identified as Missing at not Random.</p>
</div>
<br>
<div class='card-mt-4'>
    <h3>KNNImputer</h3>
    <p>The code employs the KNNImputer, a method for imputing missing values by locating the nearest neighbors for each missing value and utilizing their values as imputations. It calculates the correlations between the original and imputed values to evaluate the resemblance between the imputed and true values..</p>
    <p>The underlying concept is that the missing values can be filled in by leveraging information from other observed variables. The KNN imputation method operates on the assumption that data points in close proximity (nearest neighbors) are likely to have similar values, thereby facilitating the imputation of missing values. As the point of view, if the resulted correlation coefficient is close to one, it is the indication of <i>Missing at Random</i> type of missigness.</p>
</div>

<div style="text-align:center"><header><h1>5.2. Implementing Corresponding Data Imputations</h1></header></div>

<br>
<br>
<div class='card-body-mt-4'>
    <h3>Imputation Method for MNAR, KNNImputer</h3>
    <p>The logic behind this approach is to impute missing values in variables that have missing values using information from other observed variables. By performing KNN imputation, the code utilizes the similarity between data points (nearest neighbors) to impute the missing values. The approach assumes that similar observations will have similar values, which is leveraged to estimate the missing values. This method helps address the MNAR mechanism by incorporating relationships between variables to impute missing values.</p>
</div>
<br>
<hr>
<br>
<div class='card-body-mt-4'>
    <h3>Imputation Method for MCAR, IterativeImputer</h3>
    <p>The Iterative Imputer method uses a regression model to estimate missing values iteratively. It starts by initializing the missing values with an initial guess (e.g., mean, median, or a random value).</p> 
    <p>Then, it iteratively updates the missing values by considering each variable in turn as the target variable and using the remaining variables as predictors. The regression model is trained on the observed values and used to predict the missing values. This process is repeated multiple times until convergence, refining the imputations with each iteration.</p>
    <p>By utilizing the relationships between variables, the Iterative Imputer method captures the dependencies between variables to impute missing values more accurately than simple imputation methods. It assumes that the data is MCAR, meaning that the probability of missing values is not related to the unobserved data itself.</p>
</div>
</body>
</html>

```python
from sklearn.linear_model import LinearRegression
from sklearn.impute import KNNImputer

# Separate the variables with missing values and the complete variables
missing_vars = ['AVG_WHP_P', 'AVG_WHT_P']
other_vars = ['AVG_ANNULUS_PRESS', 'AVG_CHOKE_SIZE_P', 'DP_CHOKE_SIZE', 'AVG_DOWNHOLE_PRESSURE', 'AVG_DOWNHOLE_TEMPERATURE', 'AVG_DP_TUBING']

# Create a copy of the dataset for imputation
df_impute_3 = volve_production_producing.reset_index().copy()
df_impute_3 = df_impute_3.drop(['Years', 'DATEPRD', 'NPD_WELL_BORE_NAME', 'BORE_OIL_VOL', 'BORE_GAS_VOL', 'BORE_WAT_VOL', 'BORE_CUM_OIL', 'BORE_CUM_WAT', 'BORE_CUM_GAS', 'TOT_CUM_PRO', 'ON_STREAM_HRS'], axis=1)

# KNN imputation for other variables with missing values
imputer = KNNImputer(n_neighbors=5)
df_impute_3[other_vars] = imputer.fit_transform(df_impute_3[other_vars])

# Iterate over each missing variable
for var in missing_vars:
    # Split the data into complete and incomplete subsets
    complete_data = df_impute_3.dropna(subset=[var] + other_vars)
    incomplete_data = df_impute_3[df_impute_3[var].isnull()]

    if len(incomplete_data) > 0:  # Check if incomplete_data is not empty
        # Separate the features and target variable for imputation
        X_complete = complete_data.drop(columns=[var])
        y_complete = complete_data[var]

        # KNN imputation for the features
        imputer = KNNImputer(n_neighbors=5)
        X_complete_imputed = imputer.fit_transform(X_complete)

        # Fit a predictive model to the complete data
        model = LinearRegression()
        model.fit(X_complete_imputed, y_complete)

        # Impute the missing values using the predictive model
        incomplete_data_imputed = imputer.transform(incomplete_data.drop(columns=[var]))
        imputed_values = model.predict(incomplete_data_imputed)

        # Replace the missing values with the imputed values in the original dataset
        df_impute_3.loc[incomplete_data.index, var] = imputed_values

df_impute_3.isna().sum()
```

<h3>Cross Validation</h3>
<p>Cross-validation is a technique used to evaluate a model's performance and estimate its generalization ability on unseen data. The logic behind cross-validation is to simulate the model's performance on new data by dividing the available data into multiple subsets, typically referred to as folds. The model is trained on a subset of the data (training set) and evaluated on the remaining subset (validation set). This process is repeated multiple times, with different subsets serving as the validation set each time. The performance metrics obtained from each iteration are then averaged to provide an overall assessment of the model's performance.</p>
<p>Mathematically, cross-validation allows for a more robust estimate of a model's performance by reducing the dependency on a single train-test split. By averaging the results over multiple iterations, cross-validation provides a more reliable estimate of how well the model will perform on unseen data. It helps identify any potential issues such as overfitting or underfitting and allows for model tuning and optimization.</p>

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, classification_report
import statsmodels.api as sm
```

```python
features = flow_rates_included[['AVG_DOWNHOLE_PRESSURE', 'AVG_DP_TUBING', 'DP_CHOKE_SIZE', 'OIL_VOLUME', 'GAS_VOLUME', 'AVG_WHT_P']]
target = flow_rates_included['AVG_DOWNHOLE_TEMPERATURE']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=328)

# Define the models and their respective parameter grids
models = [
    ("Linear Regression", LinearRegression(), {}),  # Empty parameter grid for Linear Regression
    ("Random Forest", RandomForestRegressor(), {"Random Forest__n_estimators": [100, 200, 300]}),
    ("Gradient Boosting", GradientBoostingRegressor(), {"Gradient Boosting__learning_rate": [0.1, 0.05, 0.01]}),
    ("SVR", SVR(), {"SVR__C": [1, 10, 100], "SVR__epsilon": [0.1, 0.01, 0.001]}),
    ("XGBoost", XGBRegressor(), {"XGBoost__learning_rate": [0.1, 0.01, 0.001], "XGBoost__max_depth": [3, 5, 7]})
]

n_folds = 5
kf = KFold(n_splits=n_folds, shuffle=True, random_state=328)

best_models = []
best_scores = []

# Iterate over the models
for name, model, param_grid in models:
    # Create the pipeline with scaling, model, and grid search
    steps = [("scaler", StandardScaler()), (name, model)]
    pipeline = Pipeline(steps)

    # Perform cross-validation using GridSearchCV
    grid_search = GridSearchCV(pipeline, param_grid, scoring='neg_mean_squared_error', cv=kf)

    # Fit the GridSearchCV to find the best model
    grid_search.fit(X_train, y_train)

    # Get the best model and best score
    best_model = grid_search.best_estimator_
    best_score = grid_search.best_score_

    # Store the best model and best score
    best_models.append(best_model)
    best_scores.append(best_score)

# Convert the best_scores list to a numpy array and negate the values
best_scores_np = np.array(best_scores)
neg_best_scores = -best_scores_np

# Find the index of the model with the lowest RMSE
best_model_idx = np.argmin(np.sqrt(neg_best_scores))
best_model = best_models[best_model_idx]

# Print the results for the best model
print("Best Model:")
print(best_model)
best_predictions = best_model.predict(X_test)
best_rmse = np.sqrt(mean_squared_error(y_test, best_predictions))
print("Best RMSE:", best_rmse)

# Perform cross-validation on the best model
cv_scores = cross_val_score(best_model, X_train, y_train, scoring='neg_mean_squared_error', cv=kf)
cv_rmse_scores = np.sqrt(-cv_scores)
print("Cross-Validation RMSE scores:")
print(cv_rmse_scores)
print("Mean RMSE:", np.mean(cv_rmse_scores))
```

```python
# Calculate mean absolute error (MAE)
mae = mean_absolute_error(y_test, best_predictions)

# Calculate R-squared score
r2 = r2_score(y_test, best_predictions)
print("MAE:", mae)
print("R-squared:", r2)
```

```python
import plotly.express as px

# Fit the best model on the training data
best_model.fit(X_train, y_train)

# Make predictions on the validation data
y_pred = best_model.predict(X_test)

# Create a DataFrame with actual and predicted values
df_model = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

# Calculate the residuals
residuals = y_test - y_pred
df_model['Residuals'] = residuals

fig = px.scatter(df_model, x="Actual", y="Predicted", trendline="ols",
                title='Random Forest Regression Model Fit')
fig.show()
```
